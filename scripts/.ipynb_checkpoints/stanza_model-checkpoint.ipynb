```python
def convert_to_bio(doc):
    new_f = open("a.txt", "w")

    for i, sentence in enumerate(doc.sentences):
        for j, token in enumerate(sentence.tokens):
            print(token)
            start=token.start_char
            text = token.text.replace(" ", "")
            new_f.write("%s" % (text))
            new_f.write('\n')
        new_f.write('\n')
    new_f.close()
    
convert_to_bio(doc)
```

```python
def proc(filename):
    labels={}
    text=[]
    with open(filename, encoding='utf-8') as file:
        currentByte=0
        for line in file:
            cols=line.rstrip().split("\t")
            if len(cols) == 2:
                word=cols[0]
                word = word.replace(" ", "")
                label=cols[1]
                text.append(word)
                labels[currentByte]=label

                currentByte+=len(word)

    return labels, (''.join(text))


_, stanza_book = proc("book10_stanza.txt")
_, gold_book = proc("data/correct_BIO_output/book_10_BIO.txt")
```
